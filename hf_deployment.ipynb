```python
from huggingface_hub import HfApi
api = HfApi()

REPO_ID = "ML2Group7/FinalProj" 


# README first — this sets the environment
readme = (
    "---\n"
    "title: Art Style Classifier\n"
    "emoji: \U0001f3a8\n"
    "colorFrom: blue\n"
    "colorTo: purple\n"
    "sdk: gradio\n"
    "sdk_version: \"4.44.0\"\n"
    "app_file: app.py\n"
    "pinned: false\n"
    "python_version: \"3.10\"\n"
    "---\n\n"
    "# Art Style Classifier\n"
    "Upload a photo of your artwork to identify which art style it most resembles.\n"
)
with open("README.md", "w", encoding="utf-8") as f:
    f.write(readme)

# requirements — fully pinned, no ambiguity
requirements = (
    "torch==2.1.0\n"
    "torchvision==0.16.0\n"
    "timm==0.9.12\n"
    "Pillow==10.0.0\n"
    "gradio==4.44.0\n"
    "huggingface_hub==0.20.3\n"
    "pydub==0.25.1\n"
)
with open("requirements.txt", "w") as f:
    f.write(requirements)

# Upload README and requirements
for fname in ["README.md", "requirements.txt"]:
    api.upload_file(
        path_or_fileobj=fname,
        path_in_repo=fname,
        repo_id=REPO_ID,
        repo_type="space",
        commit_message=f"Fresh setup: {fname}"
    )
    print(f"✓ {fname} uploaded")

# Upload model artifacts
api.upload_file(
    path_or_fileobj="artifacts/best_model.pt",
    path_in_repo="artifacts/best_model.pt",
    repo_id=REPO_ID,
    repo_type="space",
    commit_message="Upload model weights"
)
print("✓ model uploaded")

api.upload_file(
    path_or_fileobj="data/style_labels.json",
    path_in_repo="data/style_labels.json",
    repo_id=REPO_ID,
    repo_type="space",
    commit_message="Upload label map"
)
print("✓ labels uploaded")
```
